```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]


for student in raw_students:
    # 1.Clean name
    name = student["name"].strip().title()

    # 2. Convert roll to int
    roll = int(student["roll"])

    # 3. Convert marks_str to list
    marks = student["marks_str"].split(",")
    marks = [int(m) for m in marks]

    # 4. Validate name
    is_valid = all(word.isalpha()for word in name.split())

    # 5. Print profile
    print("====================================")
    print(f"student : {name}")
    print(f"roll    : {roll}")
    print(f"marks   : {marks}")
    print("Valid Name" if is_valid else"Invalid Name")
    print("====================================")

    # 6. Roll 103 special output
    if roll == 103:
        print(name.upper())
        print(name.lower())
```

```python
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

# 1.Loop + grade
for i in range(len(subjects)):
    sub =  subjects[i]
    mark = marks[i]

    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(sub, ":", marks, "Grade:", grade)

# 2. Total & average
total = sum(marks)
avg = total / len(marks)

print("Total:", total)
print("Average:", round(avg,2))

# 3. Highest & Lowest
max_mark = max(marks)
min_mark = min(marks)

print("Highest:", subjects[marks.index(max_mark)], max_mark)
print("Lowest:", subjects[marks.index(min_mark)], min_mark)

# 4. While loop for new subjects
count = 0

while True:
    sub = input("Enter subject (or done):")

    if sub.lower() == "done":
        break

    m = input ("Enter marks:")

    if not m.isdigit():
        print("Invalid input")
        continue

    m = int(m)

    if m < 0 or m > 100:
        print("Marks must be between 0 and 100")
        continue

    subjects.append(sub)
    marks.append(m)
    count += 1

# 5. Final result
print("New subjects added:", count)

new_avg = sum(marks) / len(marks)
print("Updated average:", round(new_avg, 2))

```

```python
class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma",   [55, 68, 49, 72, 62]),
    ("Priya Nair",    [91, 85, 88, 94, 79]),
    ("Karan Mehta",   [40, 55, 38, 62, 50]),
    ("Sneha Pillai",  [75, 80, 70, 68, 85]),
]

print("Name             | Average | Status")
print("---------------------------------------")

pass_count = 0
fail_count = 0
top_avg = 0
topper = ""
total_avg = 0

for name, marks in class_data:

    avg = round(sum(marks) / len(marks), 2)

    if avg >= 60:
        status = "pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # topper check
    if avg > top_avg:
        top_avg = avg
        topper = name

    total_avg += avg

    print(f"{name:<18} | {avg:^7} | {status}")

# Final summary
print("Passed:", pass_count)
print("Failed:", fail_count)

print("Topper:", topper, top_avg)

class_avg = total_avg / len(class_data)
print("Class Average:", round(class_avg, 2))

```

```python
essay = " python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning. "

# 1. Strip
clean_essay = essay.strip()
print("Clean:", clean_essay)

# 2. Title case
print("Title Case:", clean_essay.title())

# 3. Count python
print("Count:", clean_essay.count("python"))

# 4. Replace
print("Replace:", clean_essay.replace("python", "Python 🐍"))

# 5. Split sentences
sentences = clean_essay.split(".")
print("Sentences:", sentences)

# 6. Numbered sentences
for i in range(len(sentences)):
    s = sentences[i]

    if not s.endswith("."):
        s += "."

    print(f"{i+1}. {s}")
```
